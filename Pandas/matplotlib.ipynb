```python
import pandas as pd
df = pd.read_excel(r"D:\Data science file\linechart.xlsx")
from matplotlib import pyplot as plt
plt.figure(figsize=(12,4))
plt.plot(df["Quarter"],df["Fridge"], label = "Fridge")
plt.plot(df["Quarter"],df["Dishwasher"], label = "Dishwasher")
plt.plot(df["Quarter"],df["Washing Machine"], label = "Washing machine")
plt.ylabel("Revenue (mil $)")
plt.xlabel("Financial Quarter")
plt.legend()
plt.title("SALES")
plt .show()
```

PIECHART

```python
total_sales=df[["Fridge", "Dishwasher", "Washing Machine"]].sum()
total_sales
```

```python
from matplotlib import pyplot as plt
plt.pie(total_sales, labels= total_sales.index, autopct="%1.2f%%", explode=(0.1,0,0), shadow=True)
plt.show()
```
