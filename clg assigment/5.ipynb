```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel("D:\Data science file\linechart.xlsx")

df
```

```python
plt.plot(df["Quarter"],df["Fridge"],label="Fridge")
plt.plot(df["Quarter"],df["Dishwasher"],label="Dishwasher")
plt.plot(df["Quarter"],df["Washing Machine"],label="Washing Machine")
plt.xlabel("Quarter")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
df_total=df[["Fridge","Dishwasher","Washing Machine"]].sum()
df_total
```

```python
plt.pie(df_total,labels=df_total.index,autopct="%1.2f%%",shadow=True,explode=(0.1,0,0),startangle=30)
plt.show()
```

```python
df.plot(kind="bar",x="Quarter")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```
